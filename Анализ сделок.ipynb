### Судя по датафреймам, таблицы нужно очистить и задать необходимые типы данных. Вытащить цены bid и ask, их объемы. Колонки с временем перевести datetime, остальные в int или float. Для этого в параметры read_csv укажу в качестве разделителя пробел. Получу колонки с тестом и числами, дропну текст, а с числами продолжу работу

```python
#Загрузим market логи с пробелом в качестве разделителя
#Для этого в параметрах есть delim_whitespace

mkt_logs = pd.read_csv("market_logs.log", delim_whitespace=True)
mkt_logs
```

```python
#Удалим лишние колонки, а оставшиеся переименуем 

mkt_logs = mkt_logs.drop(columns=['mLog:','market','trd','at',
                                  'with','side','price','traded',
                                  'volume','then','book','became'])

mkt_logs.columns=['trade_at','side','mkt_price','mkt_volume','after_trade']
mkt_logs
```

### Из колонки after_trade необходимо извлечь данные по ценам и объемам bid, ask. Для этого применим str.split, а в качестве параметра сначала передадим "х" в качестве разделителя, а потом "@"

```python
#Извлечем данные из колонки after_trade с разделителем "х"
#Здесь expand позволяет разделенные строки переводить сразу в колонки
#Т.е. добавятся две колонки, в которых данные будут в виде "177@1150.4"

mkt_logs[['bid','ask']] = mkt_logs['after_trade'].str.split(pat='x', expand = True)

#Из новых колонок bid, ask получаем отдельно цены и объемы
#В качестве разделителя "@"

mkt_logs[['volume_bid','bid_price']] = mkt_logs['bid'].str.split(pat='@', expand = True)
mkt_logs[['ask_price','volume_ask']] = mkt_logs['ask'].str.split(pat='@', expand = True)

#Удаляем ненужные колонки

mkt_logs = mkt_logs.drop(columns=['after_trade','bid','ask'])
mkt_logs
```

```python
#Посмотрим на типы данных в получившемся датафрейме

mkt_logs.dtypes
```

```python
#Удалим запятые из колонки mkt_price с помощью str.replace
#Колонку trade_at переведем в формат datetime, остальные - int и float

mkt_logs['trade_at'] = pd.to_datetime(mkt_logs['trade_at'])
mkt_logs['side'] = mkt_logs['side'].str.replace(',', '')
mkt_logs['mkt_price'] = mkt_logs['mkt_price'].str.replace(',', '')
mkt_logs['side'] = mkt_logs['side'].astype(float)
mkt_logs['mkt_price'] = mkt_logs['mkt_price'].astype(float)
mkt_logs['side'] = mkt_logs['side'].astype(float)
mkt_logs['volume_ask'] = mkt_logs['volume_ask'].astype(int)
mkt_logs['bid_price'] = mkt_logs['bid_price'].astype(float)
mkt_logs['ask_price'] = mkt_logs['ask_price'].astype(float)
mkt_logs.dtypes
```

```python
#Добавим колонку с mid_price

mkt_logs['mid_price'] = (mkt_logs['bid_price'] + mkt_logs['ask_price'])/2
mkt_logs
```

### Проделаем тоже самое с данными по нашим сделкам

```python
#Загрузим exec логи с пробелом в качестве разделителя

exec_logs = pd.read_csv("exec_logs.log", delim_whitespace=True)
exec_logs
```

```python
#Удалим лишние колонки, а оставшиеся переименуем

exec_logs = exec_logs.drop(columns=['eLog:','order','exec','at',
                                    'with','side','price','id',
                                    'traded','volume','volume.1',
                                    'left','on','level','delta_execsend'])

exec_logs.columns=['exec_at','id','exec_side','exec_price','exec_volume',
                   'volume_left_on_level','delta_execsend']

exec_logs
```

```python
#Типы данных в получившемся датафрейме

exec_logs.dtypes
```

```python
#Удалим запятые и зададим нужные типы данных

exec_logs['exec_at'] = pd.to_datetime(exec_logs['exec_at'])
exec_logs['id'] = exec_logs['id'].str.replace(',', '')
exec_logs['exec_side'] = exec_logs['exec_side'].str.replace(',', '')
exec_logs['exec_price'] = exec_logs['exec_price'].str.replace(',', '')
exec_logs['exec_volume'] = exec_logs['exec_volume'].str.replace(',', '')
exec_logs['volume_left_on_level'] = exec_logs['volume_left_on_level'].str.replace(',', '')
exec_logs['id'] = exec_logs['id'].astype(int)
exec_logs['exec_side'] = exec_logs['exec_side'].astype(float)
exec_logs['exec_price'] = exec_logs['exec_price'].astype(float)
exec_logs['exec_volume'] = exec_logs['exec_volume'].astype(int)
exec_logs['volume_left_on_level'] = exec_logs['volume_left_on_level'].astype(int)
exec_logs.dtypes
```

```python
#Установим в качестве индекса id сделки

exec_logs.set_index('id', inplace=True)
exec_logs
```

### Тип данных в колонке delta_execsend оставим int, т.к. это позволит нам в дальнейшем посчитать корреляцию. Corr считается только для типов данных int, float, bool. Цифры в колонке delta_execsend показывают сколько наносекунд прошло с момента отправки ордера до его исполнения. Далее для каждой сделки необходимо добавить новую колонку со значением mid_price через 5 секунд, посчитать dmid5 и корреляцию.

```python
#Зададим переменную delta в формате timedelta
#Так мы можем прибавить 5с к exec_at

delta = pd.Timedelta(5,'S')
delta
```

```python
#Создадим новую колонку с exec_at + 5с

exec_logs['exec_5s'] = exec_logs['exec_at']+delta
exec_logs
```

```python
#Создадим новый датафрейм с колонками trade_at и mid_price
#Чтобы не тащить лишние данные в exec_logs

mid_price = mkt_logs[['trade_at','mid_price']]

mid_price
```

```python
#Соединим exec_logs с mid_price с помощью merge_asof
#по колонкам exec_5s и trade_at

#Он позволит подстыковать ближайшее по времени значение mid_price
#для этого в параметрах нужно указать direction='forward'

exec_logs = pd.merge_asof(exec_logs, mid_price, 
                          left_on='exec_5s', right_on='trade_at',
                          direction='forward')
exec_logs
```

```python
#Посчитаем dmid5

exec_logs['dmid5'] = (exec_logs['exec_price']-exec_logs['mid_price'])*exec_logs['exec_side']
exec_logs
```

```python
#Посчитаем корреляцию между delta_execsend и dmid5

exec_logs['delta_execsend'].corr(exec_logs['dmid5'])
```
